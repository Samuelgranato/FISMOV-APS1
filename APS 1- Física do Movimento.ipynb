```python
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np


l=2
g=9.8

def eqdif(Y,T):
    a= Y[0]
    Va= Y[1]
    
    dadt=Va
    
    dVadt=-g*(math.sin(a))/l
    
    return [dadt,dVadt]




x=[]
y=[]
x2=[]
y2=[]

ang1=math.radians(20)
ang2=math.radians(7)


listt=np.arange(0,15,0.01) 
sol=odeint(eqdif,[ang1,0],listt)

listt=np.arange(0,15,0.01) 
sol2=odeint(eqdif,[ang2,0],listt)

for i in range(len(listt)):
    x.append(math.cos(sol[:,0][i])*l)
    y.append(math.sin(sol[:,0][i])*l)
    x2.append(math.cos(sol2[:,0][i])*l)
    y2.append(math.sin(sol2[:,0][i])*l) 

plt.plot(y,x)
plt.title("Pêndulo 20° (x por y)")
plt.ylabel('Y (m)')
plt.xlabel('X (m)')
plt.axis([min(y), max(y), max(x), min(x)])
plt.axis('equal')
plt.show()

plt.plot(y2,x2)
plt.title("Pêndulo 7° (x por y)")
plt.ylabel('Y (m)')
plt.xlabel('X (m)')
plt.axis([min(y2), max(y2), max(x2), min(x2)])
plt.axis('equal')
plt.show()



plt.plot(listt,x)
plt.title("Pêndulo 20° (x por t)")
plt.ylabel('X (m)')
plt.xlabel('T (s)')
plt.show()

plt.plot(listt,y)
plt.title("Pêndulo 20° (y por t)")
plt.ylabel('Y (m)')
plt.xlabel('T (s)')
plt.show()

plt.plot(listt,x2)
plt.title("Pêndulo 7° (x por t)")
plt.ylabel('X (m)')
plt.xlabel('T (s)')
plt.show()

plt.plot(listt,y2)
plt.title("Pêndulo 7° (y por t)")
plt.ylabel('Y (m)')
plt.xlabel('T (s)')
plt.show()





listangulo=[]
listangulo2=[]
for i in range(len(listt)):
    listangulo.append(math.degrees(sol[:,0][i]))
    listangulo2.append(math.degrees(sol2[:,0][i]))

plt.plot(listt,listangulo)
plt.title("Angulo em função do tempo 20° ")
plt.ylabel('Y (°)')
plt.xlabel('T (s)')
plt.show()

plt.plot(listt,listangulo2)
plt.title("Angulo em função do tempo 7°")
plt.ylabel('Y (°)')
plt.xlabel('T (s)')
plt.show()

```

```python
#Período:

for i in range(len(listt)):
    if listangulo[i]==min(listangulo):
        periodo=listt[i]*2
        print("O Período no lançamento de 20° é de:{0}".format(periodo))
        
    if listangulo2[i]==min(listangulo2):
        periodo2=listt[i]*2
        print("O Período no lançamento de 7° é de:{0}".format(periodo2))
```
